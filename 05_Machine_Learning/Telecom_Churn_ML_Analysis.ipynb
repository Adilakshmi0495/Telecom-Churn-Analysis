# Telecom Customer Churn Prediction
Machine Learning project to predict whether a telecom customer will churn or not.

## 1. Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Load Training and Testing Dataset
We use the 80% dataset for training and 20% dataset for testing.

```python
train_df = pd.read_csv("D:\\Telecom-Churn-Analysis\\01_Data\\churn-bigml-80.csv")
test_df = pd.read_csv("D:\\Telecom-Churn-Analysis\\01_Data\\churn-bigml-20.csv")
```

## 6. Drop Unnecessary Columns
State and Area code are not useful for prediction.

```python
train_df = train_df.drop(["State","Area code"], axis=1)
test_df = test_df.drop(["State","Area code"], axis=1)
```

## 7. Convert Categorical Columns to Numeric
Machine learning models require numerical inputs.

```python
train_df["International plan"] = train_df["International plan"].map({"Yes":1,"No":0})
train_df["Voice mail plan"] = train_df["Voice mail plan"].map({"Yes":1,"No":0})
train_df["Churn"] = train_df["Churn"].map({True:1, False:0})

test_df["International plan"] = test_df["International plan"].map({"Yes":1,"No":0})
test_df["Voice mail plan"] = test_df["Voice mail plan"].map({"Yes":1,"No":0})
test_df["Churn"] = test_df["Churn"].map({True:1, False:0})
```

## 8. Separate Features and Target Variable

```python
X_train = train_df.drop("Churn", axis=1)
y_train = train_df["Churn"]

X_test = test_df.drop("Churn", axis=1)
y_test = test_df["Churn"]
```

## 9. Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(max_iter=1000)

log_model.fit(X_train, y_train)
```

## 10. Make Predictions

```python
y_pred = log_model.predict(X_test)
```

## 11. Model Evaluation

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

## 12. Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier()

dt_model.fit(X_train, y_train)

dt_pred = dt_model.predict(X_test)

accuracy_score(y_test, dt_pred)
```

## 13. Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier()

rf_model.fit(X_train, y_train)

rf_pred = rf_model.predict(X_test)

print("confusion_matrix:")

print(confusion_matrix(y_test, rf_pred))

print("accuracy_score:", accuracy_score(y_test, rf_pred))
```

## 14. Feature Importance

```python
importance = pd.Series(rf_model.feature_importances_, index=X_train.columns)

importance.sort_values(ascending=False)
```
